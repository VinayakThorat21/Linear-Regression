```python
# Discover linear regression for math scores data
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame as df
from sklearn.linear_model import LinearRegression as linreg
```

```python
scores = pd.read_csv('lsd_math_score_data.csv')
scores.describe()
```

```python
concentration = df(scores, columns = ['LSD_ppm'])
avg_scores = df(scores, columns = ['Avg_Math_Test_Score'])
delay = df(scores, columns = ['Time_Delay_in_Minutes'])
regression = linreg()
```

```python
# Scores w.r.t. concentration
plt.title('Avearge Math Test Scores vs LSD concentration')
plt.xlabel('LSD concentration (ppm)')
plt.ylabel('Avg Score')

regression.fit(concentration, avg_scores)

# Dig deeper to extract exact value from ndarrays
slope = regression.coef_[0][0]
intercept = regression.intercept_[0]
r_squared = regression.score(concentration, avg_scores)

plt.scatter(concentration, avg_scores)
plt.plot(concentration, regression.predict(concentration), color='red', linewidth=2)
plt.show()
```

```python
# Print Stats for above visualisation
print('Slope = ' + str(round(slope, 2)))
print('Intercept = ' + str(round(intercept, 2)))
print('Goodness of fit (R-Squared) = ' + str(round(r_squared, 2)))
```

```python
# Concentration vs Time delay
# Styling
plt.style.use('classic')

plt.title('Tissue Concentration of LSD over Time', fontsize=16)
plt.ylabel('Concentration (ppm)', fontsize=14)
plt.xlabel('Time delay (min)', fontsize=14)
plt.text(s='Wagner et al. (1968)', x=320, y=1.5, fontsize=12)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.plot(delay, concentration, color='orange', linewidth=2)
plt.show()
```
